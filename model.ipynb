```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.feature_extraction.text import CountVectorizer   #To convert text into numbers
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('emails.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()  # ✅ Correct: instance
X = vectorizer.fit_transform(df['text'])  # ✅ This works now
Y=df['spam']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=10)
```

```python
model=MultinomialNB()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
score=accuracy_score(y_pred,Y_test)
print("Accuracy: ",score)
```

```python
#fun to predict if a message is spam or not

def PredictMsg(message):
    MessageVector=vectorizer.transform([message])
    prediction=model.predict(MessageVector)
    return 'Spam' if prediction[0]==1 else 'Not Spam'
#Get user input:
userMessage=input("Enter text to predict: ")
prediction=PredictMsg(userMessage)
print(f"The message is {prediction}")
```

```python
import pickle

# Save vectorizer
with open('vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)

# Save model
with open('spam_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
